```python
def get_df(data_dict):
    distance = []
    e_RA = []
    e_DE = []
    e_plx = []
    e_pmra = []
    e_pmde = []
    for key in data_dict.keys():
        distance.append(key)
        e_RA.append(data_dict[key][0])
        e_DE.append(data_dict[key][1])
        e_plx.append(data_dict[key][2])
        e_pmra.append(data_dict[key][3])
        e_pmde.append(data_dict[key][4])
    df = pd.DataFrame()
    df["Distance"] = distance
    df["e_RA"] = e_RA
    df["e_DE"] = e_DE
    df["e_plx"] = e_plx
    df["e_pmra"] = e_pmra
    df["e_pmde"] = e_pmde
    return df
```

```python
def get_grouped_df(data_task):
    distance = [(val + 1) * 100 for val in  range(100)]
    start_distance = 1
    group_dict = {}
    for dist in distance:
        end_distance = dist
        df = data_task[data_task['dist'] <= end_distance]
        df = df[df['dist'] > start_distance]
        distance = end_distance
        e_RA_median = np.nanmedian(df["ra_error"])
        e_DE_median = np.nanmedian(df["dec_error"])
        e_plx_median = np.nanmedian(df["parallax_error"])
        e_pmra_median = np.nanmedian(df["pmra_error"])
        e_pmde_median = np.nanmedian(df["pmdec_error"])
        group_dict[distance] = [e_RA_median, e_DE_median, e_plx_median, e_pmra_median, e_pmde_median]
        start_distance = dist
    
    grouped_df = get_df(group_dict)
    return grouped_df
```

```python
data_task_grouped = get_grouped_df(data_task)
data_task_grouped.head()
```
