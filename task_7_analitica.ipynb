```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import os
```

```python
with open("task_7_data\data_dr3_54") as f:
    text = f.read()
```

```python
data = pd.read_csv("task_7_data\data_dr3_54", sep=' ')
data.head()
```

```python
data.sort_values("dist", inplace = True)
data.head()
```

## 1. Parallax error

```python
plt.plot(data["dist"], data["parallax_error"])
plt.xlabel("Distance[pc]")
plt.ylabel("Parallax errors[mas]")
plt.title("Variation of parallax errors with distance")
plt.show()
```

```python
data['bins'] = [i//1000 for i in data['dist']] 
# data
```

```python
plt.hist(data['bins'])
plt.semilogy()
```

```python
data.boxplot(column=['parallax_error'], by='bins')
```

```python
files = os.listdir("task_7_data")
path = "task_7_data"
frames = [pd.read_csv(path+'/'+f, sep=' ') for f in files]
data_task = pd.concat(frames)
```

```python
data_task.head()
```

```python
data_task.sort_values("dist", inplace = True)
data_task.head()
```

```python
plt.plot(data_task.sample(n=100000)["dist"], data_task.sample(n=100000)["parallax_error"])
plt.xlabel("Distance[pc]")
plt.ylabel("Parallax errors[mas]")
plt.title("Variation of parallax errors with distance")
plt.show()
```

```python
data_task['bins'] = [i//1000 for i in data_task['dist']]
```

```python
plt.hist(data_task['bins'])
plt.semilogy()
```

```python
data_task.boxplot(column=['parallax_error'], by='bins')
```
