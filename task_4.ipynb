```python
import pandas as pd
from sys import argv
import os
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
with open("task_4_data\data_dr3_75") as f:
    text = f.read()

data = pd.read_csv("task_4_data\data_dr3_75", sep=' ')
data.head()
```

```python
plt.hist(data["phot_bp_mean_mag"], bins = 50)
plt.xlabel("Apparent BP magnitude")
plt.ylabel("Count")
plt.title("Distribution of the apparanet BP magnitude")
```
